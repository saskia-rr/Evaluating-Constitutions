```python
import pandas as pd
import json
```

```python
# Read the JSONL file
data = []
with open('agentclinic_medqa.jsonl', 'r') as file:
    for line in file:
        data.append(json.loads(line))
```

```python
patient_instructions_new = {}
for i in range(len(data)):
    point = data[i]
    diagnosis_label = point['OSCE_Examination']["Correct_Diagnosis"].replace(' ','_').lower()
    diagnosis =  point['OSCE_Examination']["Correct_Diagnosis"]
    patient = point['OSCE_Examination']["Patient_Actor"]
    patient_actor_string = f"You are a patient chatting with a doctor over an online chat interface. The doctor has never met you before. This is your profile: Demographics: {patient['Demographics']}, Overview : {patient['History']}, Primary Symptoms : {patient['Symptoms']['Primary_Symptom']}, Secondary Symptoms : {', '.join(patient['Symptoms']['Secondary_Symptoms'])}, Medical History : {patient['Past_Medical_History']}, Social History : {patient['Social_History']}, Key Review of Vitals : {patient['Review_of_Systems']}. Using the profile you should answer as the patient. Do not reveal you are an AI chatbot. Give your repsonses in 60 words or less"
    patient_instructions_new[diagnosis_label] = (diagnosis, patient_actor_string)

```

```python
patient_instructions_new
```

```python
with open("patient_instructions_new.json", "w") as outfile: 
    json.dump(patient_instructions_new, outfile)
```

```python
smaller_keys = ['hemorrhoids','endometritis','asthma','irritable_bowel_syndrome','pneumonia']
```

```python
agent_clinic_sample = {key: patient_instructions_new.get(key) for key in smaller_keys}
```
